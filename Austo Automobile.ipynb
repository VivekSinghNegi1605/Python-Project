```python
gpfl = file.loc[:,["Gender","Make","Price"]]
```

```python
gpdf = gpfl.groupby(["Gender","Make"], as_index=False)["Price"].sum()
```

```python
gpdf.Price = gpdf.Price/1000
```

```python
gpdf
```

```python
ax = sns.barplot(data=gpdf,x="Make",y="Price",hue="Gender")
ax.set_ylabel("Amount Spend in K")
ticks = [i for i in range(0,21000,2500)]
ax.set_yticks(ticks)
ax.set_yticklabels([str(t)+" K" for t in ticks])
for bars in ax.containers:
    ax.bar_label(bars)
```
